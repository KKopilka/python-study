### Бинарный поиск
##### Бинарный поиск работает по принципу «разделяй и властвуй». Он быстрее, чем линейный поиск, но требует, чтобы массив был отсортирован перед выполнением алгоритма.

```python
start_time = time.time()
def BinarySearch(mat, value):
    first = 0
    last = len(mat)-1
    index = -1
    while (first <= last) and (index == -1):
        mid = (first + last)//2
        if mat[mid] == value:
            index = mid
        else:
            if value < mat[mid]:
                last = mid - 1
            else:
                first = mid + 1
    return index

print(BinarySearch(mat, value))
print("Время бинарного поиска: ","%s seconds" % (time.time() - start_time))
```

### Фибоначчиев поиск
##### Он получил свое название потому, что использует числа Фибоначчи для вычисления размера блока или диапазона поиска на каждом шаге.

```python
start_time = time.time()
def FibonacciSearch(mat, value):
    # Два числа Фибоначчи непосредственно перед fibM
    fibM_minus_2 = 0
    fibM_minus_1 = 1
    fibM = fibM_minus_1 + fibM_minus_2
    while (fibM < len(mat)):
        fibM_minus_2 = fibM_minus_1
        fibM_minus_1 = fibM
        fibM = fibM_minus_1 + fibM_minus_2
    index = -1;
    while (fibM > 1):
        i = min(index + fibM_minus_2, (len(mat)-1))
        if (mat[i] < value):
            fibM = fibM_minus_1
            fibM_minus_1 = fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
            index = i
        elif (mat[i] > value):
            fibM = fibM_minus_2
            fibM_minus_1 = fibM_minus_1 - fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
        else :
            return i
    if (fibM_minus_1 and index < (len(mat)-1) and mat[index+1] == value):
        return index + 1;
    return -1

print(FibonacciSearch(mat, value))
print("Время фибоначчиева поиска: ","%s seconds" % (time.time() - start_time))
```

### Интерполяционный поиск
##### Интерполяционный поиск — это еще один алгоритм «разделяй и властвуй», аналогичный бинарному поиску. В отличие от бинарного поиска, он не всегда начинает поиск с середины. 

```python
start_time = time.time()
def InterpolationSearch(mat, value):
    low = 0 # Начальный индекс массива
    high = (len(mat) - 1) # Последний индекс массива
    while low <= high and value >= mat[low] and value <= mat[high]:
        # Интерполяционный поиск вычисляет вероятную позицию искомого элемента по формуле:
        index = low + int(((float(high - low) / (mat[high] - mat[low])) * (value - mat[low])))
        if mat[index] == value:
            return index
        if mat[index] < value:
            # Значение индекса пересчитывается по формуле для левого подмассива
            low = index + 1;
        else:
            # Значение индекса пересчитывается по формуле для правого подмассива
            high = index - 1;
    return -1

print(InterpolationSearch(mat, value))
print("Время интерполяционного поиска: ","%s seconds" % (time.time() - start_time))
```

### Простое рехешировние


```python
class HashTable:
    def __init__(self, n = 5):
        self.size = 0
        self.slots = [None] * n
        self.data = [None] * n
        self.q = 0

    def __setitem__(self, key, value):
        hashvalue = self._hash(key)
        if self.slots[hashvalue] is None:
            self.slots[hashvalue] = key
            self.data[hashvalue] = value
        else:
            if self.slots[hashvalue] == key:
                self.data[hashvalue] = value
            else:
                nextslot = self._rehash(hashvalue)
                self.q += 1
                while self.slots[nextslot] is not None and self.slots[nextslot] != key:
                    nextslot = self._rehash(nextslot)
                    self.q += 1
                if self.slots[nextslot] is None:
                    self.slots[nextslot] = key
                    self.data[nextslot] = value
                else:
                    self.data[nextslot] = value
        self.q = 0
    def __getitem__(self, key):
        startslot = self._hash(key)
        position = startslot
        while self.slots[position] is not None:
            if self.slots[position] == key:
                self.q = 0
                return self.data[position]
            else:
                position = self._rehash(position)
                self.q += 1
                if position == startslot:
                    continue
        self.q = 0
        return -1
    def _hash(self, key):
        return key % len(self.slots)
    def _rehash(self, oldhash):
        return (oldhash + 1) % len(self.slots)
```

### Рехэширование с помощью псевдослучайных чисел

```python
class RandomHashTable(HashTable):
    def __init__(self, n = 5):
        super().__init__(n)
        self.randints = [random.randint(0, 1000) for i in range(n)]
        
    def _rehash(self, oldhash):
        if self.q == len(self.randints):
            self.randints.extend([random.randint(0, 100) for i in range(self.q)])
        newhash = (oldhash + self.randints[self.q]) % len(self.slots)
        return newhash
```

### Метод цепочек

```python
class ListNode:
    def __init__(self, key, val):
        self.pair = (key, val)
        self.next = None

class ChainedHashTable:
    def __init__(self):
        self.m = 1000;
        self.h = [None]*self.m

    def __setitem__(self, key, value):
        index = self._hash(key)
        cur = self.h[index]
        if cur is None:
            self.h[index] = ListNode(key, value)
        else:
            while True:
                if cur.pair[0] == key:
                    cur.pair = (key, value)
                    return
                if cur.next == None: break
                cur = cur.next
            cur.next = ListNode(key, value)

    def __getitem__(self, key):
        index = self._hash(key)
        cur = self.h[index]
        while cur:
            if cur.pair[0] == key:
                return cur.pair[1]
            else:
                cur = cur.next
        return -1
        
    def _hash(self, key):
        return key % self.m
```
